# **Importing Required Libraries**

```python
import pandas as pd
import re
```

# **Loading the Dataset**

```python
df = pd.read_csv("All_Electronics.csv")

df.head()
```

# **Dataset Structure**

```python
df.columns = ['userId', 'productId', 'rating', 'timestamp']

df.head()
```

# **Data Preprocessing**

```python
product_ratings = df.groupby('productId')['rating'].mean().reset_index()

product_ratings.head()
```

# **Natural Language Query Input**

```python
query = "show electronics products with rating above 4"
```

# **Query Parsing**

```python
def extract_rating(query):

    pattern = r'rating\s*(above|over|greater than)\s*(\d)'

    match = re.search(pattern, query.lower())

    if match:
        return int(match.group(2))

    return None
```

# **Product Filtering**

```python
def search_products(query):

    rating_value = extract_rating(query)

    if rating_value:

        results = product_ratings[
            product_ratings['rating'] > rating_value
        ]

        return results.head(10)

    return product_ratings.head(10)
```

# **Example Query Execution**

```python
query = "show products with rating above 4"

results = search_products(query)

print(results)
```
